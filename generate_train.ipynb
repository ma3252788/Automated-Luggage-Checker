<h2> Purpose of this code :-</h2>

This code is used to generate train.txt , the file that contains full path for all images. During training the images will be kept in /content/Train folder in Google colab. I have written the code keeping this fact in mind.

```python
import os
```

```python
IMG_DIR = r'C:\Users\smitr\Desktop\Prohibitory Item Detection\imgs\Train'
print(os.listdir(IMG_DIR))
```

```python
all_files_list = os.listdir(IMG_DIR)

fcount = len(all_files_list)
img_list = []
i = 0

while i < fcount:
    img_list.append(all_files_list[i])
    i += 2
    
print(len(img_list))
```

```python
COLAB_DIR = '/content/Train/'

with open("train.txt",'w') as file_obj:
    for fname in img_list:
        file_obj.write(COLAB_DIR + fname)
        file_obj.write('\n')
        
file_obj.close()
```
